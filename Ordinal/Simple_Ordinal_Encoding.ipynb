```python
import pandas as pd
import datetime
```

```python
### Create The Data

data_base = datetime.datetime.today()
data_date_list = [data_base - datetime.timedelta(days=x) for x in range(0, 20)]
data = pd.DataFrame(data_date_list)
data.columns = ['day']
data
```

```python
data['day_of_week'] = data['day'].dt.day_name()
data.head()
```

```python
### Simple_Ordinal_Encoding

weekday_label = {
                'Monday':1,
               'Tuesday':2,
               'Wednesday':3,
               'Thursday':4,
               'Friday':5,
               'Saturday':6,
               'Sunday':7

}

data['encoded_data'] = data.day_of_week.map(weekday_label)
data.head(20)
```
