# Count_Or_Frequency_encoding

->If The Value_Counts Are Same This Can't Be Work.

```python
import pandas as pd
import numpy as np

data = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data' , header = None,index_col=None)
data.head()
```

```python
columns = [1,3,5,6,7,8,9,13]
```

```python
data = data[columns]
```

```python
data.columns=['Employment','Degree','Status','Designation','family_job','Race','Sex','Country']
```

```python
data.head()
```

```python
### Find The number the Lables 

for lable in data.columns[:]:
    print(lable,":",len(data[lable].unique()),"lable")
```

### Apply Count_Or_Frequency_encoding

```python
Country_map = data['Country'].value_counts().to_dict()
```

```python
data['Country'] = data['Country'].map(Country_map)
data.head(20)
```
